# Recursion vs Iteration

```python
import time

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

## Calculating factorials

```python
# iterative solution
def factorial(number):
    product = 1
    for i in range(1, number + 1):
        product *= i
    return product

iterative_times = {}
for i in range(1, 100):
    start = time.time()
    factorial(i)
    end = time.time()
    iterative_times[i] = end - start
```

```python
# recursive solution
def factorial(number):
    # base case
    if number == 1:
        return 1
    # recursive case
    else:
        return number * factorial(number - 1)

recursive_times = {}
for i in range(1, 100):
    start = time.time()
    factorial(i)
    end = time.time()
    recursive_times[i] = end - start
```

```python
sns.scatterplot(recursive_times.keys(), recursive_times.values(), label="recursive")
sns.scatterplot(iterative_times.keys(), iterative_times.values(), label="iterative")
```

## Calculating the Fibonacci sequence

```python
# iteartive solution
def fibonacci(nth_number):
    a, b = 1, 1
    for i in range(1, nth_number):
        a, b = b, a + b
    return a

iterative_times = {}
for i in range(1, 20):
    start = time.time()
    fibonacci(i)
    end = time.time()
    iterative_times[i] = end - start
```

```python
# recursive solution
def fibonacci(nth_number):
    # base case
    if nth_number == 1 or nth_number == 2:
        return 1;
    # recursive case
    else:
        return fibonacci(nth_number - 2) + fibonacci(nth_number - 1)
    
recursive_times = {}
for i in range(1, 20):
    start = time.time()
    fibonacci(i)
    end = time.time()
    recursive_times[i] = end - start
```

```python
sns.scatterplot(recursive_times.keys(), recursive_times.values(), label="recursive")
sns.scatterplot(iterative_times.keys(), iterative_times.values(), label="iterative")
```

## Calculating exponents

```python
# iterative approach
def exponent(a, n):
    result = 1
    for i in range(n):
        result *= a
    return result

iterative_times = {}
for i in range(1, 5000, 100):
    start = time.time()
    exponent(5.5, i)
    end = time.time()
    iterative_times[i] = end - start
```

```python
# recursive approach
def exponent(a, n):
    if n == 1:
        # base case
        return a
    elif n % 2 == 0:
        # recursive case (n is even)
        result = exponent(a, n // 2)
        return result * result
    elif n % 2 == 1:
        # recursive case (n is odd)
        result = exponent(a, n // 2)
        return result * result * a
    
recursive_times = {}
for i in range(1, 5000, 100):
    start = time.time()
    exponent(5.5, i)
    end = time.time()
    recursive_times[i] = end - start
```

```python
sns.scatterplot(recursive_times.keys(), recursive_times.values(), label="recursive")
sns.scatterplot(iterative_times.keys(), iterative_times.values(), label="iterative")
```
